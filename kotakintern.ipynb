## let's do label encoding for ds

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
def label_encoder(X_tr,X_tst,col_name,encoded_col_name):
    '''this function takes X_train,X_test,column name which we want to encode, and column name which we want after encoding as argument,first it encode the column, and put that encoding
    to encoded column and then it display the mapping and at the end it drop the col name'''
    label_encoder = LabelEncoder()
    encoded_tr = label_encoder.fit_transform(X_tr[col_name])
    encoded_tst = label_encoder.transform(X_tst[col_name])
    X_tr[encoded_col_name] = encoded_tr
    X_tst[encoded_col_name] = encoded_tst
    mapping = dict(zip(X_tr[col_name], encoded_tr))
    print("mapping is")
    display(mapping)
    X_tr.drop([col_name],axis=1,inplace=True)
    X_tst.drop([col_name],axis=1,inplace=True)
    
```

```python
import requests

def usd_to_inr(amount):
    url = "https://api.exchangerate-api.com/v4/latest/USD"
    response = requests.get(url)
    data = response.json()
    exchange_rate = data["rates"]["INR"]
    inr_amount = amount * exchange_rate
    return inr_amount
```

```python
print('USD 30000:',usd_to_inr(30000),'\tUSD 48000:',usd_to_inr(48000),'\tUSD 72000:',usd_to_inr(72000),'\tUSD 120000:',usd_to_inr(120000))
```

### since it is so hard to see so we will use a function which give highly correlated features and then to remove those features we will check correlation with target and lower will be removed

```python
def correlation(df,thresh):
    '''this function will give highly correlated features'''
    col_corr = list()
    corr_mat = df.corr()
    for i in range(len(corr_mat.columns)):
        for j in range(i):
            if (corr_mat.iloc[i,j])>thresh:
                col_name_1 = corr_mat.columns[i]
                col_name_2 = corr_mat.columns[j]
                col_corr.append(col_name_1)
                col_corr.append(col_name_2)
    return col_corr
```
